```python
import pandas as pd
import statsmodels.api as sm
data = pd.read_csv('ready.csv').drop(columns = 'Unnamed: 0')
data['norm_ret']= data.groupby('year_month')['month_ret'].transform(lambda x: (x - x.mean()) / x.std())
data
```

```python
def regression(data, all=True):
    if all:
        x = ['chmom', 'dolvol', 'Idiovol', 
             'indmom', 'maxret', 'mom1m', 
             'mom12m', 'mvel1', 'retvol',
             'Skew', 'RI_Spread', 'CP_Spread']
    else:
        x = ['chmom', 'dolvol', 'Idiovol', 
             'indmom', 'maxret', 'mom1m', 
             'mom12m', 'mvel1', 'retvol',]

    X = data[x]
    X = sm.add_constant(X)
    y = data['norm_ret']
    model = sm.OLS(y, X).fit()
    return model, model.ssr, model.df_resid

reduced_model, ssr_reduced, df_reduced = regression(data, all=False) #Reduced Model
full_model, ssr_full, df_full = regression(data, all=True) #Full Model

f_statistic, p_value, df_diff = full_model.compare_f_test(reduced_model)

print(f"F statistic: {f_statistic}")
print(f"P value: {p_value}")
print(f"DF Difference: {df_diff}")
```
